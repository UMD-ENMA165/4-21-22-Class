```python
## A.H. Homework 5/3/22

import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
'''necessary imports for the functions I'll use later'''

def fitting(x,y):
    plt.data('data/set1.csv')   
    
x,y = np.loadtxt('data/set1.csv', delimiter=',', unpack = True)
plt.scatter(x,y)
'''loads the data set and plots it on a scatter plot'''

def evenmore(x, a, b, c, d, e, f):
    return a * x + b * x**2 + c * x**3 + d * x**4 + e * x**5 + f

popt,pcov = curve_fit(evenmore, x, y)

xfit = np.linspace(min(x), max(x))
'''this line automatically adjusts the fitting curve to the domain of the data'''

yfit = evenmore(xfit, *popt)
plt.plot(xfit, yfit)
```
