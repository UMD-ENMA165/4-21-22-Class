```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

The mathematical equations for curve fitting

```python
def line(x, m, b):
    return m * x + b

def homework(x, a, b, c, d):
    return a *x**3 + b * x**2 + c * x + d

def trig(x, a, b, c, d):
    return a * np.sin(b - x) + c * x**2 + d

def evenmore(x, a, b, c, d, e, f):
    return a * x + b * x**2 + c * x**3 + d * x**4 + e * x**5 + f

def quadretic(x, a, b, c):
    return a * x**2 + b * x + c

def toomuch(x, a, b, c, d, e, f):
    return a * x**5 + b * x**4 + c * x**3 + d * x**2 + e * x + f

def sin(x,a,b,c):
    return a * np.sin(b * x) + c
```
