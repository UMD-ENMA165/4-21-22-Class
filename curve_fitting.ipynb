```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(0.,10.,100)
y = np.polyval([1,2,-3],x) + np.random.normal(0,10,100)
```

```python
plt.scatter(x,y)
```

```python
def line(x, m, b):
    return m * x + b
```

```python
popt,pcov = curve_fit(line, x, y)
```

```python
plt.scatter(x,y)

xfit = np.linspace(0.,10.,100)
yfit = line(xfit, popt[0], popt[1])

plt.plot(xfit, yfit)
```

```python
plt.scatter(x,y)

xfit = np.linspace(0.,10.,100)
yfit = line(xfit, *popt)

plt.plot(xfit, yfit)
```

```python
plt.scatter(x,y)

xfit = np.linspace(0.,10.,100)
mfit, bfit = popt
yfit = line(xfit, mfit, bfit)

plt.plot(xfit, yfit)

```

```python
def quadretic(x, a, b, c):
    return a * x**2 + b * x + c

popt,pcov = curve_fit(quadretic, x, y)

plt.scatter(x,y)

xfit = np.linspace(0.,10.,100)
afit, bfit, cfit =  popt
yfit = quadretic(xfit, afit, bfit, cfit)

plt.plot(xfit, yfit)

popt,pcov = curve_fit(line, x, y)

yfit = line(xfit, popt[0], popt[1])

plt.plot(xfit, yfit)
```

```python
def quadretic(x, a, b, c):
    return a * x**2 + b * x + c

def toomuch(x, a, b, c, d, e, f):
    return a * x**5 + b * x**4 + c * x**3 + d * x**2 + e * x + f

popt,pcov = curve_fit(quadretic, x, y)

plt.scatter(x,y, color='#999999')

xfit = np.linspace(0.,10.,100)
afit, bfit, cfit =  popt
yfit = quadretic(xfit, afit, bfit, cfit)

plt.plot(xfit, yfit, color = '#00dd4f', label = 'Quadretic')

popt,pcov = curve_fit(line, x, y)

yfit = line(xfit, popt[0], popt[1])

plt.plot(xfit, yfit, color = '#ff9008', label = 'Line')

popt,pcov = curve_fit(toomuch, x, y)

afit, bfit, cfit, dfit, efit, ffit = popt

yfit = toomuch(xfit, *popt)

#instructions said to have a fit for line, quadratic, and our own so I commented the 'too much' fit out
#plt.plot(xfit, yfit)
    

    
## New Fitting Curve: Exponential

def exponential(x,a,b):
    return a * b**x
popt, pcov = curve_fit(exponential, x, y)
xfit = np.linspace(0.,10.,100)
yfit = exponential(xfit, popt[0], popt[1])
plt.plot(xfit,yfit, color ='#4000ff', label = 'Exponential')
plt.legend(loc = "upper left")
```
