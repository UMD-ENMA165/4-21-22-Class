```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(0.,10.,100)
y = np.polyval([1,2,-3],x) + np.random.normal(0,10,100)
```

```python
plt.scatter(x,y)
```

```python
def line(x, m, b):
    return m * x + b
```

```python
popt,pcov = curve_fit(line, x, y)
```

```python
plt.scatter(x,y)

xfit = np.linspace(0.,10.,100)
yfit = line(xfit, popt[0], popt[1])

plt.plot(xfit, yfit)
```

```python
plt.scatter(x,y)

xfit = np.linspace(0.,10.,100)
yfit = line(xfit, *popt)

plt.plot(xfit, yfit)
```

```python
plt.scatter(x,y)

xfit = np.linspace(0.,10.,100)
mfit, bfit = popt
yfit = line(xfit, mfit, bfit)

plt.plot(xfit, yfit)
```

```python
def quad(x, a, b, c):
    return a * x**2 + b * x + c

popt,pcov = curve_fit(quad, x, y)

plt.scatter(x,y)

xfit = np.linspace(0.,10.,100)
afit, bfit, cfit =  popt
yfit = quad(xfit, afit, bfit, cfit)

plt.plot(xfit, yfit)

popt,pcov = curve_fit(line, x, y)

yfit = line(xfit, popt[0], popt[1])

plt.plot(xfit, yfit)
```

```python
def quad(x, a, b, c):
    return a * x**2 + b * x + c

def toomuch(x, a, b, c, d, e, f):
    return a * x**5 + b * x**4 + c * x**3 + d * x**2 + e * x + f

def funfunc(x, a, b):
    return np.sin(a * np.sin(b * np.sin(x)))

plt.scatter(x,y)
xfit = np.linspace(0.,10.,100)


popt,pcov = curve_fit(quad, x, y)
afit, bfit, cfit =  popt
yfit = quad(xfit, afit, bfit, cfit)
plt.plot(xfit, yfit,'g-')

popt,pcov = curve_fit(toomuch, x, y)
yfit = toomuch(xfit, *popt)
plt.plot(xfit, yfit,'r-')

popt,pcov = curve_fit(funfunc, x, y)
yfit = funfunc(xfit, popt[0], popt[1])
plt.plot(xfit, yfit, 'y-')
```
