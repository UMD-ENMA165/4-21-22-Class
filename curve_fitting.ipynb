# Curve Fitting Project

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

## Mathematical Equations

These are the mathematical equations for curve fitting

```python
def line(x, m, b):
    return m * x + b

def homework(x, a, b, c, d):
    return a *x**3 + b * x**2 + c * x + d

def trig(x, a, b, c, d):
    return a * np.sin(b - x) + c * x**2 + d

def evenmore(x, a, b, c, d, e, f):
    return a * x + b * x**2 + c * x**3 + d * x**4 + e * x**5 + f

def quadretic(x, a, b, c):
    return a * x**2 + b * x + c

def toomuch(x, a, b, c, d, e, f):
    return a * x**5 + b * x**4 + c * x**3 + d * x**2 + e * x + f

def sin(x,a,b,c):
    return a * np.sin(b * x) + c
```

## Fitting Functions

These below are functions to streamline the curve fitting and plotting code

```python
def fit_and_plot(fit_type,x,y):
    """ uses defined equations to make a graph of fit data """
    popt,pcov = curve_fit(fit_type, x, y)
    xfit = np.linspace(np.min(x),np.max(x),100)
    yfit = fit_type(xfit, *popt)
    plt.plot(xfit, yfit)

def fitting(x,y):
    """ creates a scatterplot of the fit data """
    """ uses previous function to create graphs of fit equations """
    plt.scatter(x,y)
    functions = [line, quadretic,toomuch,evenmore,trig,sin,homework]
    for equation in functions:
        fit_and_plot(equation,x,y)
```
