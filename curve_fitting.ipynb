```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(0.,10.,100)
y = np.polyval([1,2,-3],x) + np.random.normal(0,10,100)
```

```python
plt.scatter(x,y)
```

```python
def line(x, m, b):
    return m * x + b
```

```python
popt,pcov = curve_fit(line, x, y)
```

```python
plt.scatter(x,y)

xfit = np.linspace(0.,10.,100)
yfit = line(xfit, popt[0], popt[1])

plt.plot(xfit, yfit)
```

```python
plt.scatter(x,y)

xfit = np.linspace(0.,10.,100)
yfit = line(xfit, *popt)

plt.plot(xfit, yfit)
```

```python
def quadretic(x, a, b, c):
    return a * x**2 + b * x + c

popt,pcov = curve_fit(quadretic, x, y)

plt.scatter(x,y)

xfit = np.linspace(0.,10.,100)
afit, bfit, cfit =  popt
yfit = quadretic(xfit, afit, bfit, cfit)

plt.plot(xfit, yfit)

popt,pcov = curve_fit(line, x, y)

yfit = line(xfit, popt[0], popt[1])

plt.plot(xfit, yfit)
```
