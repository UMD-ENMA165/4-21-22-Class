```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
x=np.random.uniform(0.,10.,100)
y=np.polyval([1,2,-3],x)+np.random.normal(0,10,100)
plt.scatter(x,y)
def line(x,m,b):
    return m*x+b
def quad(x,a,b,c):
    return a*x**2+b*x+c
def hyper_sin(x,a,b,c):
    return a*np.sinh(x+b)+c
popt,pcov =curve_fit(line,x,y)
popt1,pcov1=curve_fit(quad,x,y)
popt2,pcov2=curve_fit(hyper_sin,x,y)
xfit=np.linspace(0.,10.,100)
plt.plot(xfit,line(xfit,popt[0],popt[1]),'r-',linewidth=3)
plt.plot(xfit,quad(xfit,popt1[0],popt1[1],popt1[2]),'m-',linewidth=3)
plt.plot(xfit,hyper_sin(xfit,popt2[0],popt2[1],popt2[2]),'y-',linewidth=3)
```
