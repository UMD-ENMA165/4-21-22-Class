```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
def line(x, m, b):
    return m * x + b
```

```python
def homework(x, a, b, c, d):
    return a *x**3 + b * x**2 + c * x + d

def trig(x, a, b, c, d):
    return a * np.sin(b - x) + c * x**2 + d

def evenmore(x, a, b, c, d, e, f):
    return a * x + b * x**2 + c * x**3 + d * x**4 + e * x**5 + f

def quadretic(x, a, b, c):
    return a * x**2 + b * x + c

def toomuch(x, a, b, c, d, e, f):
    return a * x**5 + b * x**4 + c * x**3 + d * x**2 + e * x + f

def sin(x,a,b,c):
    return a * np.sin(b * x) + c
```

```python
def fit_and_plot(fit_type,x,y):
    popt,pcov = curve_fit(fit_type, x, y)
    xfit = np.linspace(np.min(x),np.max(x),100)
    yfit = fit_type(xfit, *popt)
    plt.plot(xfit, yfit)
```

```python
x,y = np.loadtxt('data/set2.csv', delimiter=',', unpack = True)
fit_and_plot(quadretic,x,y)
```

```python
def fitting(x,y):
    plt.scatter(x,y)
   
    fit_and_plot(quadretic,x,y)
    xfit = np.linspace(np.min(x),np.max(x),100)
    
    fit_and_plot(line,x,y)

    fit_and_plot(toomuch,x,y)

    fit_and_plot(evenmore,x,y)

    fit_and_plot(line,x,y)   
    popt, pcov = curve_fit(trig, x, y)
    yfit = trig(xfit, *popt)
    plt.plot(xfit, yfit)

    popt,pcov = curve_fit(sin, x, y)
    yfit = sin(xfit, *popt)
    plt.plot(xfit, yfit)

    popt,pcov = curve_fit(homework, x, y)
    yfit = homework(xfit, *popt)
    plt.plot(xfit, yfit)
```

```python
x,y = np.loadtxt('data/set2.csv', delimiter=',', unpack = True)
fitting(x,y)
```

```python
x1,y1 = np.loadtxt('data/set1.csv', delimiter=',', unpack = True)
fitting(x1,y1)
```
