```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
my_funs=['line','quad','trig','evenmore','toomuch','sin','homework']
def line(x,m,b):
    return m*x+b
def quad(x,a,b,c):
    return a*x**2+b*x+c
def hyper_sin(x,a,b,c):
    return a*np.sinh(x+b)+c
def trig(x,a,b,c,d):
    return a*np.sin(b-x)+c*x**2+d
def evenmore(x,a,b,c,d,e,f):
    return a*x+b*x**2+c*x**3+d*x**4+e*x**5+f
def toomuch(x,a,b,c,d,e,f):
    return a*x**5+b*x**4+c*x**3+d*x**2+e*x+f
def sin(x,a,b,c):
    return a*np.sin(b*x)+c
def homework(x,a,b,c,d):
    return a*x**3+b*x**2+c*x+d
print(my_funs)
```

```python
#add fitting function hw
x=np.random.uniform(0.,10.,100)
y=np.polyval([1,2,-3],x)+np.random.normal(0,10,100)
plt.scatter(x,y)
popt,pcov =curve_fit(line,x,y)
popt1,pcov1=curve_fit(quad,x,y)
popt2,pcov2=curve_fit(hyper_sin,x,y)
xfit=np.linspace(0.,10.,100)
plt.plot(xfit,line(xfit,*popt),'r-',linewidth=3)
plt.plot(xfit,quad(xfit,*popt1),'m-',linewidth=3)
plt.plot(xfit,hyper_sin(xfit,*popt2),'y-',linewidth=3)
```
