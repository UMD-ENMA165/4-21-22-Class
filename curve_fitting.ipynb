```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(0.,10.,100)
y = np.polyval([1,2,-3],x) + np.random.normal(0,10,100)
```

```python
plt.scatter(x,y)
```

```python
def line(x, m, b):
    return m * x + b
```

```python
popt,pcov = curve_fit(line, x, y)
```

```python
plt.scatter(x,y)

xfit = np.linspace(0.,10.,100)
yfit = line(xfit, popt[0], popt[1])

plt.plot(xfit, yfit)
```

```python
def homework(x, a, b, c, d):
    return a *x**3 + b * x**2 + c * x + d

def trig(x, a, b, c, d):
    return a * np.sin(b - x) + c * x**2 + d

def evenmore(x, a, b, c, d, e, f):
    return a * x + b * x**2 + c * x**3 + d * x**4 + e * x**5 + f

def quadretic(x, a, b, c):
    return a * x**2 + b * x + c

def toomuch(x, a, b, c, d, e, f):
    return a * x**5 + b * x**4 + c * x**3 + d * x**2 + e * x + f

def sin(x,a,b,c):
    return a * np.sin(b * x) + c
```

```python
plt.scatter(x,y)

popt,pcov = curve_fit(quadretic, x, y)

xfit = np.linspace(0.,10.,100)
yfit = quadretic(xfit, *popt)

plt.plot(xfit, yfit)

popt,pcov = curve_fit(line, x, y)

yfit = line(xfit, *popt)
plt.plot(xfit, yfit)

popt,pcov = curve_fit(toomuch, x, y)

yfit = toomuch(xfit, *popt)
plt.plot(xfit, yfit)
    
popt, pcov = curve_fit(evenmore, x, y)
yfit = evenmore(xfit, *popt)

plt.plot(xfit, yfit, color = 'red')
   
popt, pcov = curve_fit(trig, x, y)
yfit = trig(xfit, *popt)

plt.plot(xfit, yfit)

popt,pcov = curve_fit(sin, x, y)
yfit = sin(xfit, *popt)

plt.plot(xfit, yfit)

popt,pcov = curve_fit(homework, x, y)

yfit = homework(xfit, *popt)
plt.plot(xfit, yfit)
```

```python
def fitting(x,y):
    plt.scatter(x,y)

    popt,pcov = curve_fit(quadretic, x, y)

    xfit = np.linspace(np.min(x),np.max(x),100)
    yfit = quadretic(xfit, *popt)

    plt.plot(xfit, yfit)

    popt,pcov = curve_fit(line, x, y)

    yfit = line(xfit, *popt)
    plt.plot(xfit, yfit)

    popt,pcov = curve_fit(toomuch, x, y)

    yfit = toomuch(xfit, *popt)
    plt.plot(xfit, yfit)
    
    popt, pcov = curve_fit(evenmore, x, y)
    yfit = evenmore(xfit, *popt)

    plt.plot(xfit, yfit, color = 'red')
   
    popt, pcov = curve_fit(trig, x, y)
    yfit = trig(xfit, *popt)

    plt.plot(xfit, yfit)

    popt,pcov = curve_fit(sin, x, y)
    yfit = sin(xfit, *popt)

    plt.plot(xfit, yfit)

    popt,pcov = curve_fit(homework, x, y)

    yfit = homework(xfit, *popt)
    plt.plot(xfit, yfit)
```
